## Tax planing
tax plan for 15%, 20%, 25%

```python
import pandas as pd
import numpy as np
```

```python
df = pd.DataFrame([{'product_id':23, 'name':'computer', 'wholesale_price': 500, 'retail_price':1000, 'sales':100},
               {'product_id':96, 'name':'Python Workout', 'wholesale_price': 35, 'retail_price':75, 'sales':1000},
               {'product_id':97, 'name':'Pandas Workout', 'wholesale_price': 35, 'retail_price':75, 'sales':500},
               {'product_id':15, 'name':'banana', 'wholesale_price': 0.5, 'retail_price':1, 'sales':200},
               {'product_id':87, 'name':'sandwich', 'wholesale_price': 3, 'retail_price':5, 'sales':300},
               ])

df
```

```python
df['current_net'] = ((-df['retail_price']+df['wholesale_price']*df['sales']))
df
```

```python
df['after_15'] = df['current_net']*0.85
df[['current_net','after_15']]
```

```python
df['after_20'] = df['current_net']*0.80
df[['current_net','after_20']]
```

```python
df['after_25'] = df['current_net']*0.75
df[['current_net','after_25']]
```

```python
df
```

```python
df['current_net'].sum() - df[['after_15','after_20','after_25']].sum()
```

25% tax on product whice cost more then 20_000

```python
tax = df['current_net'][df['current_net']>20000].sum() - (df['current_net'][df['current_net']>20000]*0.75).sum()
tax
```
