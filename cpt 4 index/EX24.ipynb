## Olympic Pivot

```python
import pandas as pd
import numpy as np
```

```python
g = np.random.default_rng(0)
df = pd.DataFrame(g.integers(0, 100, [8,3]),
columns=list('ABC'))
df['year'] = [2018] * 4 + [2019] * 4
df['quarter'] = 'Q1 Q2 Q3 Q4'.split() * 2
df
```

```python
df.pivot_table(index='quarter',
               columns= 'year',
               values= ['A', 'B', 'C'])
```

```python
df.pivot_table(index='quarter',
               columns= 'year',
               values= 'A',
               sort= 'False',
               aggfunc= 'size')
```

```python
df = pd.read_csv('../data/olympic_athlete_events.csv',
                  usecols = ['Age','Height','Team','Year','Season','Sport','Medal'],

)
df = df.loc[df['Year'] >= 1980]
df = df.loc[df['Team'].isin(['Great Britain', 'France', 'United States', 'Switzerland', 'China', 'India'])]
df
```

```python
df.pivot_table(index='Year',
               columns= 'Team',
               values= 'Age')
```

```python
df.pivot_table(index='Year',
               columns= 'Team',
               values= 'Age',
               aggfunc= 'max')
# maximum height
```

```python
df.dropna(subset= 'Medal')

df.pivot_table(index='Year',
               columns= 'Team',
               values= 'Medal',
               aggfunc= 'size')
```

```python
pd.pivot_table(df.dropna(subset='Medal'), 
               index=['Year', 'Season'], 
               columns='Team', 
               values='Medal', 
               aggfunc='size')
```

```python
df.pivot_table(index='Year', 
               columns='Team', 
               values=['Age', 'Height'], 
               aggfunc='max')
```

```python
df.pivot_table(index=['Year', 'Season'], 
            columns='Team', 
            values=['Age', 'Height'], 
            aggfunc='max')
```
