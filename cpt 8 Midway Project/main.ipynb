```python
import pandas as pd
import numpy as np
```

```python
py_df = pd.read_csv("../data/2020_sharing_data_outside.csv", 
                    low_memory= False,
                    )
py_df.head(2)
```

```python
pd.set_option('display.max_rows', 50)
pd.set_option('display.max_columns', None)
pd.options.display.float_format = '{:,.2f}'.format
```

```python
general_columns = ['age',
                   'are.you.datascientist',
                   'is.python.main',
                   'company.size',
                   'country.live',
                   'employment.status',
                   'first.learn.about.main.ide',
                   'how.often.use.main.ide',
                   'is.python.main',
                   'main.purposes'
                   'missing.features.main.ide'
                   'nps.main.ide',
                   'python.version.most',
                   'python.years',
                   'python2.version.most',
                   'python3.version.most',
                   'several.projects',
                   'team.size',
                   'use.python.most',
                   'years.of.coding'
                    ]
```

```python
def column_multi_name(column_name):
    if column_name in general_columns:
        return ('general', column_name)
    else:
        first, rest = column_name.rsplit('.',1)
        return (first, rest)
```

```python
py_df.columns = (
    pd
    .MultiIndex.from_tuples([
        column_multi_name(one_column_name)
        for one_column_name in py_df.columns
    ])
)
```

```python
py_df.head(2)
```

```python
# sorting columns
py_df = py_df[sorted(py_df.columns)]
py_df.head(2)
```

## Remove rows from so_df in which LanguageHaveWorkedWith  is NaN

```python
so_df = so_df.dropna(subset=['LanguageHaveWorkedWith'])
```

## Remove rows from so_df in whice python isn't in LanguageHaveWorkedWith

```python
so_df = so_df.loc[so_df['LanguageHaveWorkedWith'].str.contains('Python')]
```

## Remove rows from so_df in whice YearsCode NaN

```python
so_df = so_df.dropna(subset= 'YearsCode')
```

## In YeaesCode, replace "Less than 1 year" with 0 and "More then 50 years" with 51

```python
so_df.loc[so_df['YearsCode'] == "More than 50 years", 'YearsCode'] = 51
so_df.loc[so_df['YearsCode'] == "Less than 1 year", 'YearsCode'] = 0
```

```python
so_df.info()
```

```python
# turning YearsCode column into integer from object
so_df['YearsCode'] = so_df['YearsCode'].astype(np.int16)
```

## Creat a new column called "EXPERIENCE" in so_df, categorizing values in YearsCode

```python
so_df['experience'] = pd.cut(so_df['YearsCode'], 
                             bins=[-1, 1, 2, 5, 10, 100], 
                             labels=['Less than 1 year',
                             '1-2 years',
                             '3-5 years',
                             '6-10 years',
                             '11+ years'])

so_df.head()
```
