## Titanic interpolation

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("../data/titanic.csv")
df[['parch','sibsp', 'survived','pclass']] = df[['parch','sibsp','survived','pclass']].astype(np.int8)
df[['age','fare','body']] = df[['age','fare','body']].astype(np.float16)
df.head(10)
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
df.columns[df.isnull().sum() > 0 ]
```

```python
df = df.dropna(subset=['fare', 'embarked'])
df.describe()
```

```python
df['age'] = df['age'].fillna(df['age'].mean())
df.describe()
```

```python
df['home.dest'] = df['home.dest'].fillna(df['home.dest'].mode()[0])
df.describe()
```
