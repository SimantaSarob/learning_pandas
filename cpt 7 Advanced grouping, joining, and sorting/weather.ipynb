## MEAKING WEATHER DATASET MORE USABLE

```python
import pandas as pd
```

```python
d1 = '../data/san+francisco,ca.csv'
d2 = '../data/new+york,ny.csv'
d3 = '../data/springfield,ma.csv'
d4 = '../data/boston,ma.csv'
d5 = '../data/springfield,il.csv'
d6 = '../data/albany,ny.csv'
d7 = '../data/los+angeles,ca.csv'
d8 = '../data/chicago,il.csv'

df1 = pd.read_csv(d1, index_col='date_time', low_memory = False, parse_dates = ['date_time'])
df2 = pd.read_csv(d2, index_col='date_time', low_memory = False, parse_dates = ['date_time'])
df3 = pd.read_csv(d3, index_col='date_time', low_memory = False, parse_dates = ['date_time'])
df4 = pd.read_csv(d4, index_col='date_time', low_memory = False, parse_dates = ['date_time'])
df5 = pd.read_csv(d5, index_col='date_time', low_memory = False, parse_dates = ['date_time'])
df6 = pd.read_csv(d6, index_col='date_time', low_memory = False, parse_dates = ['date_time'])
df7 = pd.read_csv(d7, index_col='date_time', low_memory = False, parse_dates = ['date_time'])
df8 = pd.read_csv(d8, index_col='date_time', low_memory = False, parse_dates = ['date_time'])

df = pd.concat([df1,df2,df3,df4,df5,df6,df7,df8], axis=1)
df.head()
```

```python
df.columns = df.columns.str.replace('+','_').str.replace(',','_')
```
