```python
import pandas as pd
```

```python
oecd_df = pd.read_csv('../data/oecd_locations.csv',
                      header=None,
                      names=['abbrev', 'country'],
                      index_col='abbrev')


oecd_df.head()
```

```python
oecd_tourism_df = (
    pd
    .read_csv('../data/oecd_tourism.csv',
              usecols=['LOCATION', 'TIME', 'Value', 'SUBJECT'],
              index_col='LOCATION')
    .loc[lambda df_: df_['SUBJECT'] == 'INT-EXP']  #keeping the rows where the subject is "INT-EXP".
    .drop('SUBJECT', axis='columns') # droping the subject cause we dont need it anymore. 
)
oecd_tourism_df.head()
```

```python
wine_df = pd.read_csv('../data/winemag-150k-reviews.csv',
                      usecols=['country', 'points'])
```

```python
tourism_spending = (
    oecd_df
    .join(oecd_tourism_df)
    .groupby('country')['Value'].mean()
)

tourism_spending.head()
```

```python
country_points = wine_df.groupby('country')['points'].mean()

country_points.sort_values(ascending= False).head()
```

```python
country_points.to_frame().join(tourism_spending)
```

```python
country_points.to_frame().join(tourism_spending, how='outer')
```

```python
country_points.to_frame().join(tourism_spending, how='outer').corr()
```

```python
oecd_df_ = pd.read_csv('../data/oecd_locations.csv', 
                       header=None,
                       names=['abbrev', 
                              'country'])

oecd_tourism_df_ = pd.read_csv('../data/oecd_tourism.csv',
                              usecols=[0, 5,6],
                              header=0,
                              names=['abbrev', 
                                     'TIME', 
                                     'Value'])

wine_df_ = pd.read_csv('../data/winemag-150k-reviews.csv', 
                      usecols=['country', 
                               'points'])
```

```python
tourism_spending_ = (oecd_df
                    .merge(oecd_tourism_df_, on='abbrev')
                    .groupby('country')['Value'].mean()
                   )
tourism_spending_
```

```python
(
    country_points_.to_frame()
    .merge(tourism_spending_, on='country', how='outer')
)
```
